```python
def sigmoid(x):
    return 1 / (1 + (2.71828 ** -x))  

x1 = float(input("Enter x1: "))
x2 = float(input("Enter x2: "))

y1_expected = float(input("Enter expected y1: "))
y2_expected = float(input("Enter expected y2: "))

w1 = float(input("Enter w1: "))
w2 = float(input("Enter w2: "))
w3 = float(input("Enter w3: "))
w4 = float(input("Enter w4: "))
w5 = float(input("Enter w5: "))
w6 = float(input("Enter w6: "))
w7 = float(input("Enter w7: "))
w8 = float(input("Enter w8: "))

b1 = float(input("Enter bias b1: "))
b2 = float(input("Enter bias b2: "))

# Learning rate
alpha = float(input("Enter learning rate (alpha): "))

h1_input = (x1 * w1) + (x2 * w2) + b1
h2_input = (x1 * w3) + (x2 * w4) + b1

a1 = sigmoid(h1_input)
a2 = sigmoid(h2_input)

o1_input = (a1 * w5) + (a2 * w6) + b2
o2_input = (a1 * w7) + (a2 * w8) + b2

y1_predicted = sigmoid(o1_input)
y2_predicted = sigmoid(o2_input)

E1_derivative = -(y1_expected - y1_predicted)
E2_derivative = -(y2_expected - y2_predicted)

d_out1_netout1 = y1_predicted * (1 - y1_predicted)
d_out2_netout2 = y2_predicted * (1 - y2_predicted)

d_netout1_w5 = a1
d_netout1_w6 = a2
d_netout2_w7 = a1
d_netout2_w8 = a2

w5_new = w5 - (alpha * (E1_derivative * d_out1_netout1 * d_netout1_w5))
w6_new = w6 - (alpha * (E1_derivative * d_out1_netout1 * d_netout1_w6))
w7_new = w7 - (alpha * (E2_derivative * d_out2_netout2 * d_netout2_w7))
w8_new = w8 - (alpha * (E2_derivative * d_out2_netout2 * d_netout2_w8))


d_netout1_a1 = w5
d_netout1_a2 = w6
d_netout2_a1 = w7
d_netout2_a2 = w8

d_Etotal_a1 = (E1_derivative * d_out1_netout1 * d_netout1_a1) + (E2_derivative * d_out2_netout2 * d_netout2_a1)
d_Etotal_a2 = (E1_derivative * d_out1_netout1 * d_netout1_a2) + (E2_derivative * d_out2_netout2 * d_netout2_a2)

d_a1_net1 = a1 * (1 - a1)
d_a2_net2 = a2 * (1 - a2)

d_net1_w1 = x1
d_net1_w2 = x2
d_net2_w3 = x1
d_net2_w4 = x2

w1_new = w1 - (alpha * (d_Etotal_a1 * d_a1_net1 * d_net1_w1))
w2_new = w2 - (alpha * (d_Etotal_a1 * d_a1_net1 * d_net1_w2))
w3_new = w3 - (alpha * (d_Etotal_a2 * d_a2_net2 * d_net2_w3))
w4_new = w4 - (alpha * (d_Etotal_a2 * d_a2_net2 * d_net2_w4))

print("Updated weights:")
print(f"w1: {w1_new:.5f}, w2: {w2_new:.5f}, w3: {w3_new:.5f}, w4: {w4_new:.5f}")
print(f"w5: {w5_new:.5f}, w6: {w6_new:.5f}, w7: {w7_new:.5f}, w8: {w8_new:.5f}")
```
